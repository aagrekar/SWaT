## Data retrieval

#### Retrieving data for the first day. Samples are collected every second.

```python
db_connection = sql.connect(host='104.197.253.225', database='secdb', user='cs572', password='cs572')
db_cursor = db_connection.cursor()

#24h = 86400 samples
query = 'SELECT * FROM all_physical limit 86400'
full_data = pd.read_sql(query, db_connection)
```
